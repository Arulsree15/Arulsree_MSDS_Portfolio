```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create output folder for visuals
output_folder = "Medium 2 Visuals"
os.makedirs(output_folder, exist_ok=True)

# Load Data
file_path = "../Data/NationalDB_ChildCare_Cleaned.xlsx"
df = pd.read_excel(file_path)

# Visual 1 - Median Childcare Cost by State (Latest Year)
latest_year = df['Year'].max()
latest_data = df[df['Year'] == latest_year].groupby('State')['Median_Cost'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=latest_data.sort_values('Median_Cost', ascending=False),
            x='Median_Cost', y='State', palette='viridis')
plt.title(f"Median Childcare Cost by State - {latest_year}", fontsize=16)
plt.xlabel("Median Annual Cost (USD)")
plt.ylabel("State")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "1_cost_by_state.png"), dpi=300)
plt.show()

# Visual 2 - Cost Trend Over Time (National Average)
trend = df.groupby('Year')['Median_Cost'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(data=trend, x='Year', y='Median_Cost', marker='o', color='blue')
plt.title("National Average Childcare Cost Over Time", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Median Annual Cost (USD)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "2_national_trend.png"), dpi=300)
plt.show()

# Visual 3 - Cost vs. Median Household Income (Scatter Plot)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='MHI', y='Median_Cost', hue='Year', palette='coolwarm', alpha=0.7)
plt.title("Childcare Cost vs Median Household Income", fontsize=16)
plt.xlabel("Median Household Income (USD)")
plt.ylabel("Median Annual Cost (USD)")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "3_cost_vs_income.png"), dpi=300)
plt.show()

# Visual 4 - Average Cost by Age Group
age_group_cost = df.groupby('Age_Group')['Median_Cost'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=age_group_cost.sort_values('Median_Cost', ascending=False),
            x='Median_Cost', y='Age_Group', palette='mako')
plt.title("Average Childcare Cost by Age Group", fontsize=16)
plt.xlabel("Median Annual Cost (USD)")
plt.ylabel("Age Group")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "4_cost_by_age_group.png"), dpi=300)
plt.show()

# Visual 5 - Cost Difference by Setting
setting_cost = df.groupby('Setting')['Median_Cost'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=setting_cost.sort_values('Median_Cost', ascending=False),
            x='Median_Cost', y='Setting', palette='plasma')
plt.title("Average Childcare Cost by Setting", fontsize=16)
plt.xlabel("Median Annual Cost (USD)")
plt.ylabel("Setting")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "5_cost_by_setting.png"), dpi=300)
plt.show()

# Visual 6 - Top 10 Most Expensive States (All Years Avg)
top_states = df.groupby('State')['Median_Cost'].mean().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_states, x='Median_Cost', y='State', palette='viridis')
plt.title("Top 10 Most Expensive States for Childcare (Avg Across Years)", fontsize=16)
plt.xlabel("Median Annual Cost (USD)")
plt.ylabel("State")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "6_top10_states.png"), dpi=300)
plt.show()

print(f"✅ All visuals created and saved in '{output_folder}' folder.")
```
