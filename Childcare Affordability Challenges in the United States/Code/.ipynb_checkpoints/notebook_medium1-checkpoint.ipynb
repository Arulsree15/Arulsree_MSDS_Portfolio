```python
# 📊 CHILDCARE COSTS IN THE US - DATA ANALYSIS
# ------------------------------------------------

import pandas as pd
from IPython.display import display

# 🔹 Load the dataset
df = pd.read_excel('../Data/nationaldatabaseofchildcareprices.xlsx')

# 🔹 Preview the first few rows
print("🔍 Initial Data Preview:")
display(df.head())

# ------------------------------------------------
# 🧹 STEP 1: Select Relevant Columns for Analysis
# ------------------------------------------------

childcare_cols = ['State_Name', 'StudyYear', 'MCInfant', 'MCToddler', 'MCPreschool', 'MHI']
childcare_df = df[childcare_cols]

# ------------------------------------------------
# 📈 STEP 2: Group Data by State and Year
# ------------------------------------------------

# Calculate mean costs and income per state and year
grouped = childcare_df.groupby(['State_Name', 'StudyYear']).agg({
    'MCInfant': 'mean',
    'MCToddler': 'mean',
    'MCPreschool': 'mean',
    'MHI': 'mean'
}).reset_index()

# Rename columns for better readability
grouped.columns = [
    'State',
    'Year',
    'Avg_Infant_Cost',
    'Avg_Toddler_Cost',
    'Avg_Preschool_Cost',
    'Avg_Median_Household_Income'
]

# ------------------------------------------------
# 📋 STEP 3: Display a Sample of the Grouped Data
# ------------------------------------------------

# Select top 10 rows to display
top10 = grouped.head(10)

# Style the output as a table (displayed in the notebook)
styled_table = (
    top10.style
    .set_table_styles([
        {'selector': 'th', 'props': [('font-weight', 'bold'), ('background-color', '#f2f2f2'), ('font-size', '110%')]},
        {'selector': 'td', 'props': [('font-size', '105%')]}
    ])
    .set_caption("📌 Average Childcare Costs and Median Household Income by State and Year")
)

# Display the styled table
display(styled_table)
```

📘 Childcare Costs in the U.S.
This notebook analyzes national childcare prices based on the National Database of Childcare Prices. The goal is to understand trends, affordability, and disparities across states and time periods.

🔹 Step 1: Load and Preview the Dataset
We begin by importing the dataset and previewing the first few rows to understand the structure and types of variables included.

🔹 Step 2: Select Relevant Variables
We isolate the most important columns for our analysis—those related to childcare costs by age group and median household income.

🔹 Step 3: Group Data by State and Year
To observe trends and geographic differences, we group the data by state and year, and compute average costs and income levels.

🔹 Step 4: Preview Grouped Summary (Styled Table)
We display the top 10 rows of the cleaned, grouped dataset. The table includes average infant, toddler, preschool care costs, and median household income.

📌 Summary
This table allows us to observe variations in childcare costs and income across different states and years, and sets the stage for further analysis and visualization in later steps.
